* We will convert few columns (strings) into categorial data to apply logistic regression.

```python
pd.get_dummies(titanic['Sex']).head()
```

```python
sex=pd.get_dummies(titanic['Sex'], drop_first = True)
sex.head(3)
```

* We have dropped the first column because only one column is sufficient to determine the gender of the passenger.
Male(1) or female(0).

```python
embark = pd.get_dummies(titanic['Embarked'])
embark.head(3)
```

* 'c' stands for Cherbourg,
* 'Q' stands for Queenstown,
* 'S' stands for Southhampton

```python
embark = pd.get_dummies(titanic['Embarked'], drop_first=True)
embark.head(3)
```

* If both the values are 0 then passenger is traveling in 1st class.

```python
Pcl = pd.get_dummies(titanic['Pclass'], drop_first=True)
Pcl.head(3)
```

* Our data is now converted into categorial data.

```python
titanic = pd.concat([titanic, sex, embark, Pcl], axis=1)
titanic.head(3)
```
